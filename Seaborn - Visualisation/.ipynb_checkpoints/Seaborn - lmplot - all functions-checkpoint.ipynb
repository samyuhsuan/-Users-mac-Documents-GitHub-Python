```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
ndata = pd.read_csv("NextData.csv")
```

```python
ndata.head()
```

```python
ndata = ndata.drop(ndata.index[336::])
```

```python
ndata = ndata.drop(["Advertiser ID", "Advertiser Status", "Advertiser Integration Code"], axis=1)
```

```python
ndata.tail()
```

```python
ndata.columns
```

```python
ndata.columns=['Advertiser', 'TOD', 'Country', 'DeviceType',
       'AdvertiserCurrency', 'Impressions',
       'MeasurableImpressions',
       'ViewableImpressions', 'Clicks',
       'Spend']
```

```python
ndata.head()
```

```python
ndata["CTR"] = ndata.Clicks / ndata.Impressions
ndata["CPM"] = ndata.Spend / ndata.Impressions * 1000
ndata["Viewability"] = ndata.ViewableImpressions / ndata.MeasurableImpressions
```

```python
ndata.Country = ndata.Country.astype("category")
ndata.DeviceType = ndata.DeviceType.astype("category")
```

```python
ndata.head()
```

---

```python
#below we explore all functions except palette and set axis against lmplot
```

```python
sns.lmplot(data=ndata, x="Impressions", y="Clicks")
```

```python
#to add another dimension using colours we use hue=""
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", fit_reg=False)
```

```python
#if for some reason that you don't like the hue order you can use hue_order and pass a list 
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", 
           fit_reg=False, hue_order=["Smart Phone", "Tablet", "Desktop"])
```

```python
#to remove the linear regression line we uses fit_reg=boolean
sns.lmplot(data=ndata, x="Impressions", y="Clicks", fit_reg=False)
```

```python
#to change the size of the chart proportionally we use size=integer
#and to change the x size we use aspect=integer
sns.lmplot(data=ndata, x="Impressions", y="Clicks", fit_reg=False, size=5, aspect=2)
```

```python
#to separate the plots by the dimensions into either rows or columns
#we use row="" and col="" respectively
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", col="DeviceType", fit_reg=False)
```

```python
#markers can be used to change the markers; same as matplot
#the list of markers passable can be looked up online
#it can be a single value or a list of value 
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", col="DeviceType", fit_reg=False, 
          markers=["o", "x", "*"])
```

```python
#sharex=boolean or sharey=boolean can be used to make sure the plots are comparable or not
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", col="DeviceType", fit_reg=False, 
          sharey=False, sharex=False)
```

```python
#if there are a huge number of charts we can then wrap columns or rows using col_wrap and row_wrap repsectively
#both take integer as input
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="TOD", fit_reg=False, col="TOD", col_wrap=8)
```

```python
#if you like the clicks to impressions plot to be more aggregated then can use x_bin=integer
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", fit_reg=False, x_bins=5)
```

```python
#cosmetically we can use legend=False to remove legend 
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", fit_reg=False, legend=False)
```

```python
#or we can use legend_out=False to incorporate the legend inside the plot
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", fit_reg=False, legend_out=False)
```

```python
#lmplot automatically fit a linear line but polynominal line can be fitted when order=2 is used
#when order is larger than 2
sns.lmplot(data=ndata, x="Viewability", y="CTR", hue="DeviceType", col="DeviceType", order=2, sharey=False, sharex=False)
```

---

```python
#below we explroe using plt specific attribute: scatter_kws={} and line_kws={}
#seaborn dont have all the attributes plt has and as it's built upon plt we can use above to pass onplt attributes
#bot scatter_kws={} and line_kws={} are in the form of dictionary so use {}
```

```python
#scatter_kws={} is used to pass on scatter plot attributes from plt
#below we pass on scatter size using plt attribute "s"=integer which increases the size of the scatter
sns.lmplot(data=ndata, x="Impressions", y="Clicks", hue="DeviceType", fit_reg=False,
          scatter_kws={"s":80})
```
