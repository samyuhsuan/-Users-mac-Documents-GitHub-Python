```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ndata = pd.read_csv("NextData.csv")
```

```python
ndata.head()
```

```python
ndata = ndata.drop(ndata.index[336::])
```

```python
ndata = ndata.drop(["Advertiser ID", "Advertiser Status", "Advertiser Integration Code"], axis=1)
```

```python
ndata.columns
```

```python
ndata.columns=['Advertiser', 'TOD', 'Country', 'DeviceType',
       'AdvertiserCurrency', 'Impressions',
       'MeasurableImpressions',
       'ViewableImpressions', 'Clicks',
       'Spend']
```

```python
ndata.head()
```

```python
ndata["CTR"] = ndata.Clicks / ndata.Impressions
ndata["CPM"] = ndata.Spend / ndata.Impressions * 1000
ndata["Viewability"] = ndata.ViewableImpressions / ndata.MeasurableImpressions
```

```python
ndata.Country = ndata.Country.astype("category")
ndata.DeviceType = ndata.DeviceType.astype("category")
```

```python
ndata.head()
```

```python
#To do a heatmap we need to firstly convert it into a pivot chart before mapping it
#Below we create a pivot table broken down by TOD and DeviceType for CTR 
ndata_p = ndata.pivot_table(index="TOD", columns="DeviceType", values="CTR")
#Then we call the heatmap from sns, syntax is as below; similar to distplot o data= is needed, just strait the pivottable
sns.heatmap(ndata_p)
```

```python
#Customisation:
#--annot=True will annotate the value; default is False 
sns.heatmap(ndata_p, annot=True)
```

```python
#below we explore setting the color from specific value using:
#vmin=float, which sets the lower boundary for the color 
#vmax=float, which sets the ceiling boundary for the color 
#center=float, which sets the middle value which contrast the lower and higher value in contrast
sns.heatmap(ndata_p, annot=True, center=0.003)
```

```python
sns.heatmap(ndata_p, annot=True, vmin=0.002)
#it sets the same color for all value lower than 0.002 
```

```python
sns.heatmap(ndata_p, annot=True, vmax=0.003)
#similarly this sets the same color for all values higher than 0.003
```

```python
#dropping the bar on the righthand side
sns.heatmap(ndata_p, cbar=False)
```

```python
#adding in lines in between and specifying the line color
sns.heatmap(ndata_p, linewidths=1, linecolor="white")
```

```python
#changing the heatmap colors using the colorbrewer or error check should return all available color schemes
sns.heatmap(ndata_p, linewidths=1, linecolor="white", cmap="YlGnBu")
```

```python
#xticklabels and yticklabels can both be used to rename the plotted columns
#both of them can accept boolean: True or False which will simply be plot it or not plot it
#both can also accept list which will plot accordingly
#default is True
sns.heatmap(ndata_p, linewidths=1, linecolor="white", cmap="YlGnBu", xticklabels=False, yticklabels=False)
```

```python
#example of a list on ticklabels
sns.heatmap(ndata_p, linewidths=1, linecolor="white", cmap="YlGnBu", xticklabels=["meow", "woof", "quak"], yticklabels=False)
```

```python
#change the box sizes using a hack through subplots passing tuples through 'figsize'
f, ax=plt.subplots(1, 1, figsize=[6, 8])
sns.heatmap(ndata_p, linewidths=0.1, linecolor="white", ax=ax)
```
