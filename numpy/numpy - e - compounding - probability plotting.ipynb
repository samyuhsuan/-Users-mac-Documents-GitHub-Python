```python
#below plots the sequential performance of individual interval 
#when the interval of compounding start to increase 
```

```python
#here we create a function which receives 1 value a 
#the function plots the increase when continuously compounding at "a" number of intervals
#this funtion can demonstrate the smooth line approaching e when a starts to approach infinite
def compounding(a):
    n = np.arange(0, a)
    y = (1 + (1/(a - 1))) ** n
    plt.plot(n, y)
```

```python
compounding(100)
```

```python
#here we continue the same plotting as meow() but allows it to take a list/array of number a
#what it plots is the same above but comparison of the increase at a a list of intervals "a"
def compounding_s(a):
    for a in a:
        n = np.arange(0, a)
        y = (1 + (1/(a - 1))) ** n
        plt.plot(n, y)
```

```python
#below basically plots the sequence of each intervals from 1 interval of compouding to 100 intervals of compounding
compounding_s(np.arange(1, 100))
```

```python
#in meow1() we can see a gradual increase in y when the number of intervals start to increase 
#in meow2() we attempts the same except here we fix the x axis 
#below explains the functions:
#---for a in a it allows to take a list/array
#---initially we set n as the output of the list generated from input a
#---then the difference here comparing to meow1() is that instead of plotting n as x
#------we convert n into x through a fixed x axis - in this case 1000
#------what this allows to do is that when say n is only of 4 intervals in meow1() it will plot x = 4
#------in this case given the 1000 brackets in x axis, when n is equal to 4, x is actually an interval of 250 
#------so n is still used to plot y but against x axis instead of n being x
#---then the plot plots (x, y) intead of (n, y)
def compounding_s_fixed(a):
    for a in a:
        n = np.arange(0, a)
        x = (1000 / (a - 1)) * n
        y = (1 + (1/(a - 1))) ** n
        plt.plot(x, y)
```

```python
compounding_s_fixed(np.arange(1, 200))
```

```python
#below plots the sequential performance of probability of failing 
#the probability of failing increases when more bets are placed while the chance of winning decreases 
#n being the number of bets placed against a 1/n probability of winning
```

```python
def prob(a):
    n = np.arange(0, a)
    y = (1 - (1/(a - 1))) ** n
    plt.plot(n, y)
```

```python
#the probability of failing approaches 1/e = 0.36... when n approaches infinity
prob(100)
```

```python
#prob1() plots the increase in the probability of failing when n increases
#and put them side by side as comparison
def prob1(a):
    for a in a:
        n = np.arange(0, a)
        y = (1 - (1/(a-1))) ** n
        plt.plot(n, y)
```

```python
prob1(np.arange(1, 100))
```

```python
#prob2() is the same as prob 1 but attempts to fix the x axis
#assuming the number of intervals are fixed at 1000 intervals
def prob2(a):
    for a in a:
        n = np.arange(0, a)
        x = (1000 / (a-1) ) * n
        y = (1 - (1/(a-1))) ** n
        plt.plot(x, y)
```

```python
prob2(np.arange(1, 100))
```
